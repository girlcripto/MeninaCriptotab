### 1. Importar bibliotecas necessárias

Vamos começar importando as bibliotecas que precisaremos: `numpy` para manipulação de arrays, `matplotlib.pyplot` para visualização e `KMeans` e `make_blobs` do `sklearn` para o algoritmo K-means e geração de dados.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

### 2. Gerar dados de exemplo

Para demonstrar o K-means, vamos criar um conjunto de dados sintéticos que possua algumas 'bolhas' (clusters) naturais. Usaremos `make_blobs` para isso.

```python
# Gerar 300 amostras com 4 centros (clusters) distintos
n_samples = 300
n_features = 2
n_clusters = 4

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)

print(f"Dados gerados: {X.shape[0]} amostras com {X.shape[1]} características cada.")
```

### 3. Aplicar o algoritmo K-means

Agora vamos aplicar o K-means aos nossos dados. Definiremos o número de clusters (`n_clusters`) como 4, pois geramos os dados com essa quantidade de centros.

```python
# Inicializar e treinar o modelo K-means
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # n_init para evitar warnings
kmeans.fit(X)

# Obter os rótulos dos clusters para cada ponto de dado
labels = kmeans.labels_

# Obter as coordenadas dos centróides dos clusters
centroids = kmeans.cluster_centers_

print("K-means concluído. Centróides dos clusters:")
print(centroids)
```

### 4. Visualizar os resultados do agrupamento

Finalmente, vamos plotar os dados coloridos pelos clusters que o K-means identificou e também os centróides de cada cluster.

```python
plt.figure(figsize=(10, 7))

# Plotar cada ponto de dado, colorido de acordo com o cluster atribuído
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis', alpha=0.6)

# Plotar os centróides dos clusters
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.9, marker='X', label='Centróides')

plt.title('Agrupamento K-means dos Dados Sintéticos')
plt.xlabel('Característica 1')
plt.ylabel('Característica 2')
plt.legend()
plt.grid(True)
plt.show()
```
